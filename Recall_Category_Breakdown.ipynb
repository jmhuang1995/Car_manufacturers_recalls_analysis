```python
import matplotlib.pyplot as plt
import requests
from scipy import stats
import pandas as pd
import numpy as np
```

```python
#pulling data set
data = pd.read_csv("recall_dataset.csv")
data.head()
```

```python
#looking at what unique values sums in the data set are
data.nunique()
```

```python
#dropping columns that do not pertain to the inforation that is trying to be gathered.
data.drop(['Manufacturer Campaign','Vehicle Manufacturer'], axis = 1)
```

```python
#reviewning the columns to be able to drop
data.columns
```

```python
#dropping additional columns that are not needed.
dc_data_df = data.drop(['Manufacturer Campaign', 'Vehicle Manufacturer','Manufacture Start Date', 'Manufacture End Date',
       'Estimated Units', 'Recall Initiative', 'Recall Manufacturer',
       'Recall Notification Date'], axis = 1)
```

```python
#reviewing clean data
dc_data_df
```

```python
#pring out models from 2008 to 2018 and droping 9999 year, then seing what makes and models are left
clean_df = dc_data_df[(dc_data_df["Model Year"] > 2008) & (dc_data_df["Model Year"] != 9999)]
print(clean_df["Model Year"].unique())
print(clean_df["Vehicle Make"].unique())
```

```python
# scrubbed data one more time for only top five recalled makes and verified those years and makes 
# are the only ones left in the data set
mk_scrub_df = clean_df[(clean_df["Vehicle Make"] == "FORD") |
                      (clean_df["Vehicle Make"] == "PETERBILT") |
                      (clean_df["Vehicle Make"] == "FREIGHTLINER") | 
                      (clean_df["Vehicle Make"] == "TOYOTA") |
                      (clean_df["Vehicle Make"] == "CHEVROLET")]

print(mk_scrub_df["Vehicle Make"].unique())
print(mk_scrub_df["Model Year"].unique())
```

```python
#sorted to see them in order not really neded.
sorted_df = mk_scrub_df.groupby(["Vehicle Make","Model Year"])
```

```python
sorted_df.head()
```

```python
#scrubbed one more time to only show Recall Type "VEHICLE"
vehicle_df = mk_scrub_df[(mk_scrub_df['Recall Type'] == 'VEHICLE')]
```

```python
vehicle_df
```

```python
#Scrubbed to only get fords info to able to make a list of unique recalls for a list
Ford_df = vehicle_df.loc[(vehicle_df["Vehicle Make"] == "FORD") ]
Equipment_Recalls_df = Ford_df.loc[(Ford_df["Recall Component"].str.contains("EQUIPMENT"))]
Drp_Dup_Equipement_Recalls_df = Equipment_Recalls_df.drop_duplicates(subset=["Recall Component"])
Drp_Dup_Equipement_Recalls_df


        
        
```

```python
#created for loop to make list of only recalls numbers to be able to call to and API
ford_recall = []
for index, row in Drp_Dup_Equipement_Recalls_df.iterrows():
    ford_recall.append(row["NHTSA Campaign"])
ford_recall
```

```python
#api request to pull detailed info of what the recall is
url = ("https://api.nhtsa.gov/recalls/campaignNumber?campaignNumber=")
ford_recall
for recall in ford_recall:
        response = requests.get(url + recall).json()
        print("______")
        print(response["results"][0]["Summary"])
    
```
